```python
code_master = pd.read_excel('sample_codemaster.xlsx')
code_master
```

```python
sample_1_code = pd.merge(left = sample_1,
                        right = code_master,
                        how = 'left',
                        left_on = '국적코드',
                        right_on = '국적코드')
sample_1_code
```

```python
sample_1_code2 = pd.merge(left = sample_1,
                        right = code_master,
                        how = 'inner',
                        left_on = '국적코드',
                        right_on = '국적코드')
sample_1_code2
```

```python
sample_2 = pd.read_excel("sample_2.xlsx",
                        header = 1,
                        skipfooter = 2,
                        usecols = "A:C")
sample_2['기준년월'] = '2019-12'
```

```python
sample_2_code = pd.merge(left = sample_2,
                        right = code_master,
                        how = 'left',
                        left_on = '국적코드',
                        right_on = '국적코드')
sample_2_code
```

```python
sample_1_code
```

```python
sample = sample_1_code.append(sample_2_code,ignore_index=True)
sample
```

```python
sample_F = sample_1_code.append(sample_2_code,ignore_index=False)
sample_F
```

```python
sample.to_excel("sample.xlsx")
```
